```python
# =========================================================
# KLASIFIKASI CITRA DENGAN CNN:
# DARI SCRATCH HINGGA TRANSFER LEARNING
# DATASET: CIFAR-10
# =========================================================

# =========================================================
# IMPORT LIBRARY
# =========================================================
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import time
import warnings

from sklearn.metrics import (
    confusion_matrix,
    classification_report,
    precision_score,
    recall_score,
    f1_score,
    roc_curve,
    auc
)

from sklearn.preprocessing import label_binarize
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA

warnings.filterwarnings('ignore')

print("TensorFlow Version:", tf.__version__)
print("GPU Available:", tf.config.list_physical_devices('GPU'))

# =========================================================
# LOAD DATASET CIFAR-10
# =========================================================
print("\nLOADING CIFAR-10 DATASET...")
print("=" * 50)

(X_train, y_train), (X_test, y_test) = keras.datasets.cifar10.load_data()

print("Training Shape :", X_train.shape)
print("Testing Shape  :", X_test.shape)

# =========================================================
# NORMALISASI
# =========================================================
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# =========================================================
# NAMA KELAS
# =========================================================
class_names = [
    'airplane',
    'automobile',
    'bird',
    'cat',
    'deer',
    'dog',
    'frog',
    'horse',
    'ship',
    'truck'
]

# =========================================================
# ONE HOT ENCODING
# =========================================================
num_classes = 10

y_train_cat = keras.utils.to_categorical(y_train, num_classes)
y_test_cat = keras.utils.to_categorical(y_test, num_classes)

# =========================================================
# VISUALISASI DATASET
# =========================================================
plt.figure(figsize=(10,10))

for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(X_train[i])
    plt.title(class_names[y_train[i][0]])
    plt.axis('off')

plt.suptitle("Sample CIFAR-10 Images", fontsize=16)
plt.tight_layout()
plt.show()

# =========================================================
# DATA AUGMENTATION
# =========================================================
print("\nMEMBUAT DATA AUGMENTATION...")
print("=" * 50)

datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    zoom_range=0.2,
    shear_range=0.2,
    fill_mode='nearest'
)

datagen.fit(X_train)

# =========================================================
# VISUALISASI AUGMENTASI
# =========================================================
plt.figure(figsize=(12,6))

sample = X_train[0]

for i in range(10):
    augmented = datagen.random_transform(sample)

    plt.subplot(2,5,i+1)
    plt.imshow(np.clip(augmented,0,1))
    plt.axis('off')

plt.suptitle("Data Augmentation Examples")
plt.tight_layout()
plt.show()

# =========================================================
# CNN FROM SCRATCH
# =========================================================
print("\nMEMBANGUN CNN FROM SCRATCH...")
print("=" * 50)

def build_cnn_model(optimizer='adam'):

    model = Sequential([

        # BLOCK 1
        layers.Conv2D(
            32,
            (3,3),
            activation='relu',
            padding='same',
            input_shape=(32,32,3)
        ),

        layers.BatchNormalization(),

        layers.Conv2D(
            32,
            (3,3),
            activation='relu',
            padding='same'
        ),

        layers.MaxPooling2D((2,2)),
        layers.Dropout(0.25),

        # BLOCK 2
        layers.Conv2D(
            64,
            (3,3),
            activation='relu',
            padding='same'
        ),

        layers.BatchNormalization(),

        layers.Conv2D(
            64,
            (3,3),
            activation='relu',
            padding='same'
        ),

        layers.MaxPooling2D((2,2)),
        layers.Dropout(0.25),

        # BLOCK 3
        layers.Conv2D(
            128,
            (3,3),
            activation='relu',
            padding='same'
        ),

        layers.BatchNormalization(),

        layers.Conv2D(
            128,
            (3,3),
            activation='relu',
            padding='same'
        ),

        layers.MaxPooling2D((2,2)),
        layers.Dropout(0.25),

        # FULLY CONNECTED
        layers.Flatten(),

        layers.Dense(
            256,
            activation='relu'
        ),

        layers.Dropout(0.5),

        layers.Dense(
            num_classes,
            activation='softmax'
        )
    ])

    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# =========================================================
# MODEL ADAM
# =========================================================
cnn_adam = build_cnn_model('adam')

cnn_adam.summary()

# =========================================================
# CALLBACKS
# =========================================================
callbacks = [

    keras.callbacks.EarlyStopping(
        monitor='val_accuracy',
        patience=8,
        restore_best_weights=True
    ),

    keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=4,
        min_lr=1e-6
    )
]

# =========================================================
# TRAINING CNN
# =========================================================
print("\nTRAINING CNN WITH ADAM...")
print("=" * 50)

start_train = time.time()

history_adam = cnn_adam.fit(
    datagen.flow(X_train, y_train_cat, batch_size=64),
    epochs=30,
    validation_data=(X_test, y_test_cat),
    callbacks=callbacks,
    verbose=1
)

end_train = time.time()

training_time = end_train - start_train

print(f"\nTraining Time: {training_time:.2f} seconds")

# =========================================================
# EVALUASI CNN
# =========================================================
print("\nEVALUASI CNN...")
print("=" * 50)

start_infer = time.time()

pred_probs = cnn_adam.predict(X_test)

end_infer = time.time()

inference_time = end_infer - start_infer

pred_classes = np.argmax(pred_probs, axis=1)
true_classes = y_test.flatten()

accuracy = np.mean(pred_classes == true_classes)

precision = precision_score(
    true_classes,
    pred_classes,
    average='weighted'
)

recall = recall_score(
    true_classes,
    pred_classes,
    average='weighted'
)

f1 = f1_score(
    true_classes,
    pred_classes,
    average='weighted'
)

print(f"Accuracy       : {accuracy:.4f}")
print(f"Precision      : {precision:.4f}")
print(f"Recall         : {recall:.4f}")
print(f"F1-Score       : {f1:.4f}")
print(f"Inference Time : {inference_time:.4f} sec")

# =========================================================
# LEARNING CURVE
# =========================================================
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(14,5))

# ACCURACY
ax1.plot(history_adam.history['accuracy'], label='Train Accuracy')
ax1.plot(history_adam.history['val_accuracy'], label='Validation Accuracy')

ax1.set_title("Accuracy Curve")
ax1.set_xlabel("Epoch")
ax1.set_ylabel("Accuracy")
ax1.legend()
ax1.grid(True)

# LOSS
ax2.plot(history_adam.history['loss'], label='Train Loss')
ax2.plot(history_adam.history['val_loss'], label='Validation Loss')

ax2.set_title("Loss Curve")
ax2.set_xlabel("Epoch")
ax2.set_ylabel("Loss")
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()

# =========================================================
# CONFUSION MATRIX
# =========================================================
cm = confusion_matrix(true_classes, pred_classes)

plt.figure(figsize=(10,8))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=class_names,
    yticklabels=class_names
)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# =========================================================
# CLASSIFICATION REPORT
# =========================================================
print("\nCLASSIFICATION REPORT")
print("=" * 50)

print(classification_report(
    true_classes,
    pred_classes,
    target_names=class_names
))

# =========================================================
# ROC CURVE MULTICLASS
# =========================================================
print("\nMEMBUAT ROC CURVE...")
print("=" * 50)

y_test_bin = label_binarize(true_classes, classes=np.arange(num_classes))

plt.figure(figsize=(10,8))

for i in range(num_classes):

    fpr, tpr, _ = roc_curve(
        y_test_bin[:, i],
        pred_probs[:, i]
    )

    roc_auc = auc(fpr, tpr)

    plt.plot(
        fpr,
        tpr,
        label=f'{class_names[i]} (AUC={roc_auc:.2f})'
    )

plt.plot([0,1],[0,1],'k--')

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Multiclass")
plt.legend()
plt.grid(True)

plt.show()

# =========================================================
# VISUALISASI PREDIKSI
# =========================================================
plt.figure(figsize=(12,12))

for i in range(25):

    plt.subplot(5,5,i+1)

    plt.imshow(X_test[i])

    pred_label = class_names[pred_classes[i]]
    true_label = class_names[true_classes[i]]

    color = 'green'

    if pred_classes[i] != true_classes[i]:
        color = 'red'

    plt.title(
        f"T:{true_label}\nP:{pred_label}",
        color=color,
        fontsize=8
    )

    plt.axis('off')

plt.suptitle("Prediction Results")
plt.tight_layout()
plt.show()

# =========================================================
# FEATURE MAP VISUALIZATION
# =========================================================
print("\nVISUALISASI FEATURE MAP...")
print("=" * 50)

feature_model = keras.Model(
    inputs=cnn_adam.inputs,
    outputs=cnn_adam.layers[0].output
)

sample_image = X_test[0:1]

feature_maps = feature_model.predict(sample_image)

plt.figure(figsize=(12,8))

for i in range(16):

    plt.subplot(4,4,i+1)

    plt.imshow(feature_maps[0,:,:,i], cmap='viridis')

    plt.axis('off')

plt.suptitle("Feature Maps")
plt.tight_layout()
plt.show()

# =========================================================
# FILTER VISUALIZATION
# =========================================================
print("\nVISUALISASI FILTER...")
print("=" * 50)

filters, biases = cnn_adam.layers[0].get_weights()

plt.figure(figsize=(10,10))

for i in range(16):

    f = filters[:,:,:,i]

    f_min, f_max = f.min(), f.max()

    f = (f - f_min) / (f_max - f_min)

    plt.subplot(4,4,i+1)

    plt.imshow(f)

    plt.axis('off')

plt.suptitle("Conv Filters")
plt.tight_layout()
plt.show()

# =========================================================
# PCA VISUALIZATION
# =========================================================
print("\nPCA VISUALIZATION...")
print("=" * 50)

feature_extractor = keras.Model(
    inputs=cnn_adam.inputs,
    outputs=cnn_adam.layers[-3].output
)

features = feature_extractor.predict(X_test[:2000])

pca = PCA(n_components=2)

pca_result = pca.fit_transform(features)

plt.figure(figsize=(10,8))

scatter = plt.scatter(
    pca_result[:,0],
    pca_result[:,1],
    c=true_classes[:2000],
    cmap='tab10',
    alpha=0.7
)

plt.colorbar(scatter)

plt.title("PCA Feature Embedding")
plt.show()

# =========================================================
# t-SNE VISUALIZATION
# =========================================================
print("\nt-SNE VISUALIZATION...")
print("=" * 50)

tsne = TSNE(
    n_components=2,
    random_state=42,
    perplexity=30
)

tsne_result = tsne.fit_transform(features[:1000])

plt.figure(figsize=(10,8))

scatter = plt.scatter(
    tsne_result[:,0],
    tsne_result[:,1],
    c=true_classes[:1000],
    cmap='tab10',
    alpha=0.7
)

plt.colorbar(scatter)

plt.title("t-SNE Feature Embedding")
plt.show()

# =========================================================
# TRANSFER LEARNING
# =========================================================
print("\nTRANSFER LEARNING")
print("=" * 50)

# =========================================================
# RESIZE CIFAR-10 UNTUK PRETRAINED MODEL
# =========================================================
X_train_resize = tf.image.resize(X_train, (96,96))
X_test_resize = tf.image.resize(X_test, (96,96))

# =========================================================
# FUNCTION TRANSFER LEARNING
# =========================================================
def build_transfer_model(base_model):

    base_model.trainable = False

    model = Sequential([

        base_model,

        layers.GlobalAveragePooling2D(),

        layers.Dense(
            256,
            activation='relu'
        ),

        layers.Dropout(0.5),

        layers.Dense(
            num_classes,
            activation='softmax'
        )
    ])

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )

    return model

# =========================================================
# VGG16
# =========================================================
print("\nTRAINING VGG16...")
print("=" * 50)

vgg_base = keras.applications.VGG16(
    weights='imagenet',
    include_top=False,
    input_shape=(96,96,3)
)

vgg_model = build_transfer_model(vgg_base)

history_vgg = vgg_model.fit(
    X_train_resize,
    y_train_cat,
    epochs=5,
    batch_size=64,
    validation_data=(X_test_resize, y_test_cat),
    verbose=1
)

vgg_loss, vgg_acc = vgg_model.evaluate(
    X_test_resize,
    y_test_cat,
    verbose=0
)

print(f"VGG16 Accuracy: {vgg_acc:.4f}")

# =========================================================
# RESNET50
# =========================================================
print("\nTRAINING RESNET50...")
print("=" * 50)

resnet_base = keras.applications.ResNet50(
    weights='imagenet',
    include_top=False,
    input_shape=(96,96,3)
)

resnet_model = build_transfer_model(resnet_base)

history_resnet = resnet_model.fit(
    X_train_resize,
    y_train_cat,
    epochs=5,
    batch_size=64,
    validation_data=(X_test_resize, y_test_cat),
    verbose=1
)

resnet_loss, resnet_acc = resnet_model.evaluate(
    X_test_resize,
    y_test_cat,
    verbose=0
)

print(f"ResNet50 Accuracy: {resnet_acc:.4f}")

# =========================================================
# MOBILENETV2
# =========================================================
print("\nTRAINING MOBILENETV2...")
print("=" * 50)

mobilenet_base = keras.applications.MobileNetV2(
    weights='imagenet',
    include_top=False,
    input_shape=(96,96,3)
)

mobilenet_model = build_transfer_model(mobilenet_base)

history_mobile = mobilenet_model.fit(
    X_train_resize,
    y_train_cat,
    epochs=5,
    batch_size=64,
    validation_data=(X_test_resize, y_test_cat),
    verbose=1
)

mobile_loss, mobile_acc = mobilenet_model.evaluate(
    X_test_resize,
    y_test_cat,
    verbose=0
)

print(f"MobileNetV2 Accuracy: {mobile_acc:.4f}")

# =========================================================
# PERBANDINGAN MODEL
# =========================================================
print("\nPERBANDINGAN MODEL")
print("=" * 50)

models_name = [
    'CNN Scratch',
    'VGG16',
    'ResNet50',
    'MobileNetV2'
]

accuracies = [
    accuracy,
    vgg_acc,
    resnet_acc,
    mobile_acc
]

plt.figure(figsize=(10,6))

bars = plt.bar(
    models_name,
    accuracies
)

plt.ylabel("Accuracy")
plt.title("Model Comparison")

for bar, acc in zip(bars, accuracies):

    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        f"{acc:.3f}",
        ha='center'
    )

plt.show()

# =========================================================
# FINE TUNING
# =========================================================
print("\nFINE TUNING MOBILENETV2...")
print("=" * 50)

mobilenet_base.trainable = True

for layer in mobilenet_base.layers[:-20]:
    layer.trainable = False

mobilenet_model.compile(
    optimizer=keras.optimizers.Adam(1e-5),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_finetune = mobilenet_model.fit(
    X_train_resize,
    y_train_cat,
    epochs=5,
    batch_size=64,
    validation_data=(X_test_resize, y_test_cat),
    verbose=1
)

fine_loss, fine_acc = mobilenet_model.evaluate(
    X_test_resize,
    y_test_cat,
    verbose=0
)

print(f"\nAccuracy Setelah Fine-Tuning: {fine_acc:.4f}")

# =========================================================
# KESIMPULAN HASIL
# =========================================================
print("\nKESIMPULAN")
print("=" * 50)

print(f"""
1. CNN From Scratch Accuracy  : {accuracy:.4f}

2. VGG16 Accuracy             : {vgg_acc:.4f}

3. ResNet50 Accuracy          : {resnet_acc:.4f}

4. MobileNetV2 Accuracy       : {mobile_acc:.4f}

5. Fine-Tuning Accuracy       : {fine_acc:.4f}

6. Training Time              : {training_time:.2f} detik

7. Inference Time             : {inference_time:.4f} detik

Kesimpulan:
- Transfer Learning umumnya memberikan accuracy lebih tinggi.
- Data augmentation membantu mengurangi overfitting.
- Fine-tuning meningkatkan performa model pretrained.
- MobileNetV2 lebih ringan dan cepat.
- ResNet50 biasanya memberikan accuracy terbaik.
""")
```
